```python
from PIL import Image
from matplotlib.pyplot import imshow
from numpy import *
from scipy.ndimage import measurements,morphology
from scipy.misc import imsave
import imageio
```

```python
#load image convert to greyscale and threshold
path1 = (r'C:\Users\Reaper124\TestScripts\morphology\coins_2.jpg')
im = array(Image.open(path1).convert('L'))
imshow(np.asarray(im))
imageio.imwrite('grey_coins_2.jpg',im)
#threshold image for all pixels < 128
im = 1*(im<128)
labels, nbr_objects = measurements.label(im)
print("number of objects:", nbr_objects)
```

```python
#morphology - opening to seperate objects better
imshow(np.asarray(im))
im_open = morphology.binary_opening(im,ones((9,7)),iterations=2)

labels_open, nbr_objects_open = measurements.label(im_open)
print("Number of objects:", nbr_objects_open)
```
